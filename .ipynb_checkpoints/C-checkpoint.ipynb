* GlobalID: 每个目击者记录的唯一标签
* Detection Data: 报告的观测日期
* Notes: 由提交报告的人提供的评论。这可以是一名公众成员，偶尔也可以是一名国家雇员
* Lab Status: 经过分析后，美国农业部正式对其进行了分类。Positive ID确定是亚洲大黄蜂。Negative ID意味着它被排除在外。  
  Unprocessed(未处理)的意思是它还没有被分类。Unverified(未经核实)的意思是由于缺乏信息而没有做出决定。
* Lab Comments: 国家昆虫学实验室分析后添加到记录中的内容。
* Submission Date: 向国家提交报告的日期，此日期可显著后测日期。
* Latitude(of sighting): 经度
* Longitude(of sighting): 纬度

不能用Lab Comments这个变量，因为这是分析后的人说的话，而不是分析前的话

```python
# 对其他格式转图片格式的图片编码
import base64
import os
# img_path = r'..\数据\原数据\14只大黄蜂的图片\ATT2_HoireRDBeekeeperReport.jpg'

img_paths = os.listdir(path='.\other_to_jpg')
img_baseDic = {}
for im_path in img_paths:
    with open('.\other_to_jpg\\'+im_path, 'rb') as f:
        image_data = f.read()
        base64_data = base64.b64encode(image_data)  # base64编码
        img_baseDic[im_path] = str(base64_data, encoding='utf-8')
# 警告：请不要查看img_baseDic的数据，会内存不足！
# img_baseDic数据
```

```python
# 计算 F1-Measure，当系数β=1
TP = cm_te[0][0]
TN = cm_te[0][1]
FP = cm_te[1][0]
FN = cm_te[1][1]
# 精确率 precision
P = TP/(TP+FP)
# 召回率 recall
R = TP/(TP+FN)
# 综合评价指标
F = (2*P*R)/(P+R)
print('P:',P, 'R:', R, 'F:', F)
```

###### 失败的评分模型

```python
def cal_scale(score,odds,PDO):
    '''
    参数：
        odds: 设定的坏好比
        score: 在这个odds下的分数
        PDO： 好坏翻倍比
        model: 逻辑回归模型
    return：
        A,B
    '''
    B = PDO/np.log(2)
    A = score + B*np.log(odds)
    # base_score = A+B*model.intercept_[0]
    print('B: {:.2f}'.format(B))
    print('A: {:.2f}'.format(A))
    # print('基础分为：{:.2f}'.format(base_score))
    return A,B

def Pre2Score(pre,A,B):
    y = np.log(pre/(1-pre))
    return round(1300-A+B*y) # 这里是用1300- 因为概率越高分数越高
```

```python
A,B = cal_scale(770,0.05,13) # 假设基础分为50，odds为5%，PDO为10
se['score'] = se['lr_pro'].map((lambda x:Pre2Score(x,A,B)))
# se['score'].describe()
# se['lr_pro'].map((lambda x:Pre2Score(x,A,B))).describe()
```
