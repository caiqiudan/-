##### 收纳

```r
rm <- read.csv('datas/model_data.csv', sep=',',encoding="UTF-8")
```

```r
y = rm[43]      # 已对数化
x = rm[1:42]

model_data <- cbind(y, x)
```

```r
result <- lm(model_data)
summary(result)
```

###### 逐步回归分析

```r
tstep <- step(result)
summary(tstep)
```

```r
tstep
```

```r
# 主成分分析
x.pr <- princomp(x, cor=TRUE)
summary(x.pr, loadings=TRUE)
```

##### AIC准则

```r
rm <- read.csv('../复赛/datas/modelData.csv', sep=',',encoding="UTF-8")
```

```r
colnames(rm)
```

```r
library(leaps)
```

```r
# y = rm[43]      # 已对数化
# y = exp(y)      # 未对数化
# x = rm[1:42]

# model_data <- cbind(y, x)

# # 重命名列名
# colnames(model_data)[1] <- 'y'
```

AIC

```r
exps <- regsubsets(y~., data=rm, nbest=1, really.big=T)
expres <- summary(exps)

res_aic <- data.frame(expres$outmat, adjr2=expres$adjr2)
# res_aic;       # 查看结果(看不清)
```

```r
plot(exps,scale="adjr2")
```

```r
write.csv(res_aic, file='..\\复赛\\datas\\最优子集_初始版aic.csv')
```

Cp

```r
res_cp <- data.frame(expres$outmat, Cp=expres$cp)
```

```r
write.csv(res_cp, file='..\\复赛\\datas\\最优子集_初始版cp.csv')
```

BIC

```r
res_bic <- data.frame(expres$outmat, BIC=expres$bic)
```

```r
write.csv(res_bic, file='..\\复赛\\datas\\最优子集_初始版bic.csv')
```
