### 数据整合

#### 周贡榜的信息合并

```python
wc = pd.read_csv('data\周贡献榜.csv')
wc.用户id = wc.用户id.astype(int)
# wc.drop(columns=['iSFFlag','sLogo','tNobleLevel'], inplace=True)   # 删除一些没有用信息的列
```

字段概念定义：  

|  字段   | 定义  |
|  ----  | ----  |
| 金主  | 贡献排名前5%为金主 |
| 广泛贡献用户  | 对5个以上的主播打赏过的用户 |
| 高等级用户  | 16级以上的用户(全部主播的贡献用户的前20%) |

```python
# 贡献榜总人数
a = wc.groupby('用户id').lUid.count().rename('贡献榜总人数').reset_index()
pht = pd.merge(pht, a, on='用户id', how='left')
pht['贡献榜总人数'].fillna(0, inplace=True)
```

```python
# 是否有收入  
a = wc.drop_duplicates('lUid').lUid.values
pht['是否有收入'] = pht.用户id.isin(a).map({True:1, False:0})
pht['是否有收入'] = pht.贡献榜总金额.apply(lambda x: 1 if x>0 else 0)
```

```python
# 贡献榜总金额
a = wc.groupby('用户id').iScore.sum().rename('贡献榜总金额').reset_index()
pht = pd.merge(pht, a, on='用户id', how='left')
pht['贡献榜总金额'].fillna(0, inplace=True)
```

```python
# 贡献榜平均等级
a = wc.groupby('用户id').iUserLevel.mean().rename('贡献榜平均用户等级').reset_index()
pht = pd.merge(pht, a, on='用户id', how='left')
# pht['贡献榜平均用户等级'].fillna(-1, inplace=True)   # 缺失值
```

```python
# 贵宾数
wc.loc[wc[wc.iNobleLevel==0].index , 'iNobleLevel'] = np.nan   # 守护者等级为nan的表示不是守护者，
a = wc.groupby('用户id').iNobleLevel.count().rename('贵宾数').reset_index()
pht = pd.merge(pht, a, on='用户id', how='left')
pht['贵宾数'].fillna(0, inplace=True)
```

```python
# 广泛贡献用户数
a = wc.groupby('lUid').sNickName.count(); a = a[a>3].index
wc['是否为广泛贡献用户'] = wc.lUid.isin(a).map({True:1, False:0})
a = wc.groupby('用户id').是否为广泛贡献用户.sum().rename('广泛贡献用户数').reset_index()
pht = pd.merge(pht, a, on='用户id', how='left')
pht['广泛贡献用户数'].fillna(0, inplace=True)
```

```python
# 金主数量
# 是否为金主
a = wc.iScore.sort_values(ascending=False)
wc['是否金主'] = wc.index.isin(a[:int(len(a)/5)].index)
wc['是否金主'] = wc['是否金主'].map({True:1, False:0})

a = wc.groupby('用户id').是否金主.sum().rename('金主数量')
pht = pd.merge(pht, a, on='用户id', how='left')
pht['金主数量'].fillna(0, inplace=True)
```

```python
# 高等级用户数
wc.loc[wc[wc.iUserLevel>16].index, '是否为高等级用户'] = 1
a = wc.groupby('用户id').是否为高等级用户.sum().rename('高等级用户数')
pht = pd.merge(pht, a, on='用户id', how='left')
pht['高等级用户数'].fillna(0, inplace=True)
```

```python
# 高等级用户占比
pht['高等级用户数占比'] = (pht['高等级用户数']/pht['贡献榜总人数']).fillna(0)
```

```python
# 守护者个数
a = wc.groupby('用户id').lUid.count().rename('守护者个数')
pht = pd.merge(pht, a, on='用户id', how='left')
```

```python
# 贡献榜人均金额
pht['贡献榜人均金额'] = (pht['贡献榜总金额']/pht['贡献榜总人数']).fillna(0)
```

#### 视频动态的信息合并

```python
iv = pd.read_csv('data\视频动态.csv')
iv.drop_duplicates(inplace=True)
iv['cTime'] = iv.cTime.astype('datetime64')
iv = iv.rename(columns={'uid':'用户id'})    # 重命名uid
import time
iv['cTime'] = iv.cTime.map(time.localtime).map(lambda x: time.strftime("%Y--%m--%d %H:%M:%S", x)).astype('datetime64')
# 提取字段
iv['观看次数'] = iv.videoInfo.apply(lambda x: eval(x)['videoPlayNum'])
# 观看时长
start_time = datetime.strptime('1900-1-1', '%Y-%m-%d')
iv['视频时长'] = iv.videoInfo.apply(lambda x: eval(x)['videoDuration'])
iv['视频时长'] = iv.视频时长.apply(lambda x: datetime.strptime(x, '%M:%S') if len(x)==5 else datetime.strptime(x, '%H:%M:%S'))
iv['视频时长'] = iv.视频时长.apply(lambda x: (x - start_time).seconds)
```

```python
# 视频平均点赞数  
a = iv.groupby('用户id').favorCount.mean().rename('视频平均点赞数')
pht = pd.merge(pht, a, on='用户id', how='left')
```

```python
# 视频平均评论数  
a = iv.groupby('用户id').commentCount.mean().rename('视频平均评论数')
pht = pd.merge(pht, a, on='用户id', how='left')
```

```python
# 视频平均分享数  
a = iv.groupby('用户id').shareCount.mean().rename('视频平均分享数')
pht = pd.merge(pht, a, on='用户id', how='left')
```

```python
# 视频平均时长  
a = iv.groupby('用户id').视频时长.mean().rename('视频平均时长')
pht = pd.merge(pht, a, on='用户id', how='left')
```

```python
# 视频平均观看次数
a = iv.groupby('用户id').观看次数.mean().rename('视频平均观看次数')
pht = pd.merge(pht, a, on='用户id', how='left')
```

```python
# 每周视频更新频率
a = iv.groupby('用户id')[['cTime']].resample(rule='W', on='cTime', how='count').groupby(level=0).mean()  # 运行时间较长
b = a.groupby(level=0).mean().rename(columns={'cTime':'视频平均每周更新频率'})
pht = pd.merge(pht, b, on='用户id', how='left')
```

##### 在播时段：  
  
    14:00 - 01:00 黄金时段  大于均值  
    01:00 - 14:00 非黄金时段  小于均值  

```python
def get_hour(t):
    return t.hour if t else np.nan
pht['上半天直播开始时间点'] = pht.上半天直播开始时间.map(get_hour).fillna(-1)  # -1填充，方便下一步的运算
pht['下半天直播开始时间点'] = pht.下半天直播开始时间.map(get_hour).fillna(-1)

pht['开始直播时间点'] = pht[['上半天直播开始时间点','下半天直播开始时间点']].apply(lambda x: x[0] if x[0]>x[1] else x[1], axis=1)
```

```python
bins = [0, 1, 14, 25]
labels = ['黄金时段_2', '非黄金时段',  '黄金时段']
pht['活跃时间'] = pd.cut(pht.开始直播时间点, bins=bins, labels=labels, right=False).apply(lambda x: x[:4])
```

##### 公告主题——LDA

```python
def cut2list(x):
    try:
        return list(x)
    except:
        return ['']
    
r = '[’!"#$%&\'()*+,-./|:|：|;<=>?@[\\]^_`{|}~]+|欢迎|来到|直播间|的|、|小|不|群|便|客|'    # 去掉符号
r_ty = re.sub('\n','|',a)[955:]                                       # 去掉停用词 我、你、就等
# r = r+r_ty
pht.公告.fillna('', inplace=True)
pht['公告词袋'] = pht.公告.apply(lambda x: re.sub(r_ty, '', re.sub(r,'', x))).map(jieba.cut).map(cut2list)
```

```python
# 2.生成词频和语料
texts = pht.公告词袋.to_list() # 评论数最高的企业进行分析
dictionary = corpora.Dictionary(texts)                 # 生成词频字典
corpus = [dictionary.doc2bow(text) for text in texts]  # 生成语料库
```

```python
# 3.LDA模型的训练
num_topics = 3
lda = models.LdaModel(corpus=corpus, id2word=dictionary, num_topics=3, iterations=500, random_state=100)  # 迭代200次
```

```python
# 查看各主题的主题词  
word_pro = lda.get_topic_terms(0, topn=10000)
word_pro_x = list(map(lambda x: x[0],word_pro))
word_pro_y = list(map(lambda x: x[1],word_pro))

topic_list = lda.print_topics(num_words=40) # num_words：显示词数
df_topic = pd.DataFrame()
len_df = 29
for n,topic in enumerate(topic_list):
#     print(topic)
    df_topic.insert(0,f'topic{n}',re.findall('\*"(\w+)"', topic[1])[:len_df])

# 查看每个主题前num_words个词
df_topic
# 主题1：直播时间
# 主题2：求订阅、求关注
# 主题3：粉丝马甲、卡牌子
```

```python
# lda模型预测  
# 测试数据转换
test_vec = [dictionary.doc2bow(doc) for doc in pht.公告词袋]
#预测并打印结果
for n, item in enumerate(test_vec):
    topic = lda.get_document_topics(item)
    a = [i for i in [j[1] for j in topic]]
    pht.loc[n, '公告主题'] = a.index(max(a)) + 1

titleDic = {1:'直播时间', 2:'求订阅', 3:'粉丝马甲'}
pht['公告主题'] = pht.公告主题.map(titleDic)
```

```python
pht['直播标题词袋'] = pht.直播标题.apply(lambda x: re.sub(r_ty, '', re.sub(r,'', x))).map(jieba.cut).map(cut2list)
# 2.生成词频和语料
texts = pht.直播标题词袋.to_list() # 评论数最高的企业进行分析
dictionary = corpora.Dictionary(texts)                 # 生成词频字典
corpus = [dictionary.doc2bow(text) for text in texts]  # 生成语料库
# 3.LDA模型的训练
num_topics = 3
lda_title = models.LdaModel(corpus=corpus, id2word=dictionary, num_topics=3, iterations=500, random_state=100)  # 迭代200次
```

```python
# 查看各主题的主题词  
word_pro = lda.get_topic_terms(0, topn=10000)
word_pro_x = list(map(lambda x: x[0],word_pro))
word_pro_y = list(map(lambda x: x[1],word_pro))

topic_list_title = lda_title.print_topics(num_words=40) # num_words：显示词数
df_topic = pd.DataFrame()
len_df = 29
for n,topic in enumerate(topic_list_title):
#     print(topic)
    df_topic.insert(0,f'topic{n}',re.findall('\*"(\w+)"', topic[1])[:len_df])

# 查看每个主题前num_words个词
df_topic
# 主题1：鱼塘乱杀
# 主题2：大神教学
# 主题3：妹纸陪玩
```
