```python
def cut2list(x):
    try:
        return list(x)
    except:
        return ['']
    
r = '[!"#$%&\'()*+,-./|:||;<=>?@[\\]^_`{|}~]+|||||||||||'    # 
r_ty = re.sub('\n','|',a)[955:]                                       #  
# r = r+r_ty
pht..fillna('', inplace=True)
pht[''] = pht..apply(lambda x: re.sub(r_ty, '', re.sub(r,'', x))).map(jieba.cut).map(cut2list)
```

```python
# 2.
texts = pht..to_list() # 
dictionary = corpora.Dictionary(texts)                 # 
corpus = [dictionary.doc2bow(text) for text in texts]  # 
```

```python
# 3.LDA
num_topics = 3
lda = models.LdaModel(corpus=corpus, id2word=dictionary, num_topics=3, iterations=500, random_state=100)  # 200
```

```python
#   
word_pro = lda.get_topic_terms(0, topn=10000)
word_pro_x = list(map(lambda x: x[0],word_pro))
word_pro_y = list(map(lambda x: x[1],word_pro))

topic_list = lda.print_topics(num_words=40) # num_words
df_topic = pd.DataFrame()
len_df = 29
for n,topic in enumerate(topic_list):
#     print(topic)
    df_topic.insert(0,f'topic{n}',re.findall('\*"(\w+)"', topic[1])[:len_df])

# num_words
df_topic
# 1
# 2
# 3
```

```python
# lda  
# 
test_vec = [dictionary.doc2bow(doc) for doc in pht.]
#
for n, item in enumerate(test_vec):
    topic = lda.get_document_topics(item)
    a = [i for i in [j[1] for j in topic]]
    pht.loc[n, ''] = a.index(max(a)) + 1

titleDic = {1:'', 2:'', 3:''}
pht[''] = pht..map(titleDic)
```

```python
pht..value_counts()
```

```python
text_cut = []
pht..apply(lambda x: text_cut.append(x))
text_cut = ' '.join([' '.join(i) for i in text_cut])
text_cut = re.sub('|','',text_cut)
```

```python
from wordcloud import WordCloud
from wordcloud import ImageColorGenerator
from PIL import Image                            # 
# text_cut = 'recommend go use money someth 3 button stop work month use...' # text_cut

background = Image.open("image/2.png")
graph = np.array(background)
pic_color=ImageColorGenerator(graph)   #

word_cloud = WordCloud(scale=4,background_color="white",  #   # scale=4,
                       mask=graph, font_path='simfang.ttf',    #  ;   'simsun' 
                      max_words=500, color_func=pic_color,   # max_words     max_words=2000,  
                      max_font_size=100)                  # 
word_cloud.generate(text_cut)
plt.subplots(figsize=(10,8), dpi=150)
plt.imshow(word_cloud)
plt.axis("off")
plt.savefig('.png', format='png', bbox_inches='tight',  # 
                dpi=300, transparent=True)
```
